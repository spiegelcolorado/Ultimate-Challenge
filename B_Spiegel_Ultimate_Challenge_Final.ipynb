```python
ult_df_30 =ult_df[ult_df['last_trip_date'] > '2014-06-01']
print(ult_df_30.shape)
df_inactive = ult_df[ult_df['last_trip_date'] < '2014-06-01']
df_inactive.shape
df_inactive['active'] = 0
df_active = ult_df_30
df_active['active'] = 1
type(df_active)
```

```python


frames = [df_active, df_inactive]

result = pd.concat(frames)
```

```python
result = result.dropna(how = 'any', axis = 0)
result.info()
result_dum = pd.get_dummies(result.city)
result_dm = pd.concat([result, result_dum ], axis = 1)
```

```python
result_dm = result_dm.drop('city', axis = 1)
```

```python
phone_dic = {'iPhone' :1, 'Android' : 2}
result_dm['phone'] = result_dm['phone'].map(phone_dic)
```

# Retaining Customers
40.5 percent of users were retained over a 6 month time peiod

```python
X = result_dm.drop(['active','last_trip_date'], axis = 1)
y = result_dm['active']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(y_train.shape)
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(X_train, y_train)
RF_pred = RFC.predict(X_test)
RF_SCORE = RFC.score(X_test, y_test)

```

```python
RF_SCORE
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, RF_pred)
```
